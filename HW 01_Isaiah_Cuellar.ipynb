(5 points) Generate 5 random numbers whose values are between 1 and 20. However, skip the number if the random number generated is equal to 13. You must use loops (either for loop or while loop) to solve this problem.

You should still print 5 random numbers even if you skip the number 13.

```python
#Title HW1 Isaiah Cuellar
#Problem 1
import random 

i = 1
while i < 6:
  some_integer = random.randint(1, 20)
  if (some_integer == 13):
    #print("We hit a 13!")
    continue
  i += 1
  print(some_integer)
```

(10 points)  Write a python script to generate multiplication tables for the numbers 5 to 9. And each multiplication table should go from 1 to 10. You must use loops (either for loop or while loop) to solve this problem.

For example, your first multiplication table will be as follows

5 x 1 = 5

5 x 2 = 10

...

5 x 10 = 50

And your last multiplication table will be as follows.

9 x 1 = 9

9 x 2 = 18

...

9 x 10 = 90

```python
#Problem 2

for x in range (5, 10): #Outter Loop, loops from 5-9
  for y in range (1, 11): #Inner loop, loops from 1-10
    total = x * y
    print(str(x) + " x " + str(y) + " = " + str(total) + "\n")
```

(10 points) Write a script that calculates the squares and cubes of the numbers from 1 to 5. You must use loops (either for loop or while loop) to solve this problem.

Print the resulting values in a table format, as shown below. Use the tab escape sequence to achieve the three-column output.

number               square                 cube

1                            1                            1            

2                            4                            8

3                            9                            27

4                            16                          64

5                             25                          125

```python
#Problem 3

def square_root(number):
  return number ** 2

def cubed(number):
  return number ** 3

def print_all(number):
  print(str(number) + "\t\t" + str(square_root(number)) + "\t\t" + str(cubed(number)) + "\n")

print("Number" + "\t\tSqaure" + "\t\tCube")
for x in range(1, 6):
  print_all(x)

```
